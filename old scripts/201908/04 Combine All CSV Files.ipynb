```python
#set to the file extension of "to-be-merged" files
ext = ".csv"
#set to your working directory
dir_path = 'c://PersonalOneDrive//OneDrive//Work//Census//Python Scripts//201908'
#set to the name of your output file
#results = 'ACSSF_2012_2016_Full.txt'

files = os.listdir('c://PersonalOneDrive//OneDrive//Work//Census//Python Scripts//201908')

#e20161ak0001000
#e = estimates
#2016 = year
#1 = 1 year estimate vs. 5 year
#ak = state
#0001 = sequence number
#000 = iterationID

files2 = [k for k in files if 'csvfi' in k]
#files3 = [k for k in files2 if 'e2016' in k]
#files3 = [k for k in files2 if '0041' in k]


string = 'c://PersonalOneDrive//OneDrive//Work//Census//Python Scripts//201908//'
files4 = [string + x for x in files2]
files4
```

```python
logfile = pd.read_csv("c://PersonalOneDrive//OneDrive//Work//Census//Python Scripts//201908//LogRecReference.csv", engine='python')
#logfile = logfile.drop(['source','Unnamed: 0','Name'], 1)

logfile1 = logfile[logfile['State'].str.contains('MA')]
logfile2 = logfile[logfile['State'].str.contains('CT')]
logfile3 = logfile[logfile['State'].str.contains('VT')]
logfile4 = logfile[logfile['State'].str.contains('NH')]
logfile5 = logfile[logfile['State'].str.contains('RI')]
logfile6 = logfile[logfile['State'].str.contains('ME')]
#logfile6 = logfile[logfile['LOGRECNO']==1]

logfile = pd.concat([logfile6,logfile5,logfile4, logfile3,logfile2,logfile1], ignore_index=True)

logfile.rename(columns = {"Logical Record Number":"LOGRECNO", "State":"STATE"}, inplace = True)

logfile.head()
```

```python
for f in files4:
  if f.endswith(ext):
    data = open(f)
    
     
    inputcsv = pd.read_csv(data,  engine='python')
    
    #inputcsv = inputcsv.drop(['4','1','Unnamed: 0','GEOID','Name','YEAR'], 1)

    #	Unnamed: 0			STATE			LOGRECNO
    inputcsv = inputcsv.drop(['FILEID','FILETYPE','Unnamed: 0','CHARITER'], 1)
    
    inputcsv.rename(columns = {"SEQUENCE":"SEQ"}, inplace = True)
    
    
    
    inputcsv.rename(index=str, columns= {'SEQ' : f})

    
    newcolumn = inputcsv['SEQ'].values
    seq_num = str(newcolumn[0])[0:3]
    new_Seq_columnname = "SEQ" + seq_num + "_"   

    columnnames = inputcsv.columns[~inputcsv.columns.str.contains('STATE|SEQ|LOGRECNO')]
    inputcsv.rename(columns = dict(zip(columnnames, new_Seq_columnname + columnnames)),inplace=True)
    
    
    logfile = pd.merge(logfile, inputcsv, on = ['LOGRECNO', 'STATE'], how = "left")    
    
    #newcolumn = logfile['SEQ'].values

    #seq_num = str(newcolumn[0])[0:3]
    #seq_num = seq_num[0: seq_num.find(".")]
    New_Seq_Title = "SEQ_"+ seq_num
        
    
    logfile.rename(columns = {'SEQ' : New_Seq_Title}, inplace=True)
    
    data.close()
    
   
```
