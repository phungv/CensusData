```python
columnnames = pd.read_csv("\\\\Aopscifs004.amsa.com\\fs$\\DATA\\GROUP\\Insights and Outcomes\\IOG Projects\\MA Congressional District Map\\Data\\FilteredFileHeaders_Hamp.csv", engine='python')
#columnnames = pd.read_excel(Seqfile, sheet_name='E')
values = {'FILEID': "1", 'FILETYPE': "YEAR", 'STUSAB': "STATE",'CHARITER': "4","SEQUENCE" : "SEQ", "LOGRECNO": "LOGRECNO"}

columnnames2 = columnnames.fillna(value = values)
name_str = columnnames2.transpose() 
name_str = name_str.drop([1],1)
name_str = name_str.drop([2],1)
#name_str

Filteredfile.columns = name_str.values
Filteredfile.columns = Filteredfile.columns.map(str)
Filteredfile.columns = Filteredfile.columns.str.strip().str.replace(":", '').str.replace("'", '').str.replace(',', '').str.replace('(', '').str.replace(')', '')

Filteredfile

CleanedFilteredfile = Filteredfile.drop(['x'],1)


sheet_names = ['AK', 'AL', 'AR', 'AZ', 'CA', 'CO', 'CT', 'DC', 'DE',
 'FL', 'GA', 'HI', 'IA', 'ID', 'IL', 'IN', 'KS', 'KY', 'LA', 'MA', 'MD', 'ME',
 'MI', 'MN', 'MO', 'MS', 'MT', 'NC', 'ND', 'NE', 'NH', 'NJ', 'NM', 'NV', 'NY',
 'OH', 'OK', 'OR', 'PA', 'PR', 'RI', 'SC', 'SD', 'TN', 'TX', 'US',
 'UT', 'VA', 'VT', 'WA', 'WI', 'WV', 'WY']

dfs = pd.read_excel("C:\\Census Data\\References\\5yearMiniGeo.xlsx", sheet_name=sheet_names)

LogRecRef = pd.concat((df.assign(source=sheet) for sheet, df in dfs.items()), ignore_index=True)


CleanedFilteredfile = pd.merge(CleanedFilteredfile, LogRecRef, on = ['LOGRECNO', 'STATE'], how = "left")
CleanedFilteredfile = CleanedFilteredfile.drop(['source','GEOID_y'], 1)
    
CleanedFilteredfile.rename(columns = {'GEOID_x' : 'GEOID'}, inplace=True)
CleanedFilteredfile.rename(columns = {'Name' : 'Location Name'}, inplace=True)

CleanedFilteredfile    
```
